Можем заметить, что количество регионов на 1 меньше действительного. Это может быть связано с тем, что в одном регионе находятся два субъекта с различными значениями в полях okato и oktmo. Нужно найти этот двойственный регион. С 2014 ОКТМО заменил ОКАТО, поэтому будем ориентироваться на него.

```python
regions = dict()
for i in range(train.shape[0]):
    if train.region[i] not in regions.keys():
        regions[train.region[i]] = set()
        regions[train.region[i]].add(train.oktmo[i])
    else:
        regions[train.region[i]].add(train.oktmo[i])
regions[21]
```
