```python
regions = dict()
for i in range(train.shape[0]):
    if train.region[i] not in regions.keys():
        regions[train.region[i]] = set()
        regions[train.region[i]].add(train.oktmo[i])
    else:
        regions[train.region[i]].add(train.oktmo[i])
[i for i in regions.values() if len(i) > 1]
```

## Дата

Ни один из алгоритмов машинного обучения не умеет работать со столбцом даты напрямую, поэтому нужно извлечь все данные в отдельные столбцы.

```python
train.insert(4, 'year', train.date.dt.year)
train.insert(5, 'month', train.date.dt.month)
train.insert(6, 'day', train.date.dt.day)
train.insert(7, 'day_of_week', train.date.dt.dayofweek)
```

# Построение пайплайна

```python
features = ['region', 'year', 'month', 'day', 'day_of_week']
y_pasta = train.pasta
X_pasta = pd.get_dummies(train[features], columns=['region', 'month', 'day', 'day_of_week', 'year'])
X_pasta
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error

X_pasta_train, X_pasta_test, y_pasta_train, y_pasta_test = train_test_split(
                            X_pasta, y_pasta, test_size=0.9, random_state=0)
# column_encoding = make_column_transformer((OneHotEncoder(), features))
logreg = LogisticRegression(solver='lbfgs')
```

```python
# pipe = make_pipeline(column_encoding, logreg)
```

```python
logreg.fit(X_pasta_train, y_pasta_train)
```

```python
prediction = logreg.predict(X_pasta_test)
```

```python
prediction
```

```python
mean_squared_error(y_pasta_test, prediction.astype('float64'))
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_pasta_test, prediction.astype('float64'))
```
