## Исправление формата числовых значений


Все количесвенные данные представлены в виде строк, более того, в качестве разделителя используется запятая вместо точки. Исправим это в тренировочном датасете.

```python
numeric_columns = train.columns.drop(['region', 'oktmo', 'okato', 'date'])
numeric_columns
```

В процессе нашлась еще одна пробелема - пробелы ('xa0') во многих столбцах, решил ее с помощью replace. 

```python
for column in numeric_columns:
    for i in range(train.shape[0]):
        value = train[f'{column}'][i].replace(',', '.').strip()
        value = value.replace(u'\xa0', '')
        train.at[i, f'{column}'] = float(value)
```

```python
train[numeric_columns] = train[numeric_columns].astype('float64')
```

```python
train.dtypes
```

# Загрузка дополнительных данных

Не запрещается использовать дополнительную информацию из открытых источников, чтобы попробовать увеличить точность модели.

```python
from datetime import timedelta

def fill_missing_values(df_with_missing_values, start_value, diff_dates):
    d_full = dict()
    d_full['2019-01-01 00:00:00'] = start_value
    for date in datelist:
        if date in diff_dates:
            d_full[f'{date}'] = d_full[f'{date - timedelta(days=1)}']
        else:
            d_full[f'{date}'] = float(df_with_missing_values[df_with_missing_values.Date == date].Price)
    return d_full    
```

```python
data_date = set(train.date) | set(test.date)
```

```python
datelist = pd.date_range('2019-01-02', '2021-06-30', freq="d")
datelist
```

## Курс доллара

Идея проста - предсказуемые величины могут зависеть от курса доллара ЦБ, установленного на каждый конкретный день. Поэтому я решил загрузить [данные](http://www.cbr.ru/currency_base/dynamics/) по курсу доллара на каждый рассматриваемый день и использовать его при построении модели.

```python
dollar_df = pd.read_excel('dollar.xlsx')
dollar_df.drop('Unnamed: 0', axis=1, inplace=True)
dollar_df.rename(columns={'data' : 'Date', 'curs' : 'Price'}, inplace=True)
```

```python
dollar_df
```

```python
dollar_date = set(dollar_df.Date)
diff_dates_dollar = data_date - dollar_date
```

```python
dollar = fill_missing_values(dollar_df, 69.4706, diff_dates_dollar)
usd = pd.Series(dollar, name='usdrub')
usd
```

```python
usd_train = usd[:'2021-03-31 00:00:00'].sort_index(ascending=False)
usd_test = usd['2021-04-01 00:00:00':].sort_index(ascending=False)
```

```python
usd_train, usd_test
```

## Индекс РТС (RTSI)

Логика та же, что и с долларом. Будем брать значение закрытия для каждого дня из нашего [датасета](https://www.finanz.ru/indeksi/arhiv-torgov/rts).

```python
rts_df = pd.read_csv('RTS.csv', sep=';')
rts_df.drop(['Open', 'DailyHigh', 'DailyLow'], axis=1, inplace=True)
rts_df.rename(columns={'ClosingPrice' : 'Price'}, inplace=True)
```

```python
for i in range(rts_df.shape[0]):
    value = float(rts_df['Price'][i].replace(',', '.'))
    rts_df.at[i, 'Price'] = value
```

```python
rts_df.Price = rts_df.Price.astype('float64')
rts_df.Date = pd.to_datetime(rts_df.Date, format='%d.%m.%Y')
```

```python
rts_df
```

```python
rts_date = set(rts_df.Date)
diff_dates_rts = data_date - rts_date
```

```python
rts = fill_missing_values(rts_df, 1066.13, diff_dates_rts) 
rtsi = pd.Series(rts, name='rtsi')
rtsi
```

```python
rtsi_train = rtsi[:'2021-03-31 00:00:00'].sort_index(ascending=False)
rtsi_test = rtsi['2021-04-01 00:00:00':].sort_index(ascending=False)
```

```python
rtsi_train, rtsi_test
```

# Создание и тренировка моделей

```python
y = train[numeric_columns]
y.columns
```

```python
features = ['region', 'month', 'day', 'day_of_week', 'usdrub', 'rtsi', 'brent']

X = pd.get_dummies(train[features])
X
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=0)
model = RandomForestRegressor(n_estimators=100, min_samples_split=5, random_state=0, n_jobs=-1)
```

```python
model.fit(X_train, y_train)
```

```python
importances = model.feature_importances_
importances
```

```python
indices = np.argsort(importances)[::-1]
names = ['region', 'brent', 'month', 'usdrub', 'rtsi', 'day_of_week', 'day']
```

```python
indices
```

```python
plt.figure()
plt.title('Feature importance')
plt.bar(range(len(names)), importances[indices])
plt.xticks(range(len(names)), names, rotation=90)
plt.show()
```

```python
prediction = model.predict(X_valid)
```

```python
pred = dict()
j = 0
for column in numeric_columns:
    values = list()
    for i in range(len(prediction)):
        values.append(prediction[i][j])
    j += 1
    pred[f'{column}'] = np.array(values)
```

```python
mean = list()
mae = list()

for column in numeric_columns:
    mean.append(y_valid[f'{column}'].mean())
    mae.append(mean_absolute_error(y_valid[f'{column}'], pred[f'{column}']))    
```

```python
E = list()

for i in range(len(mae)):
    E.append(mae[i] / mean[i])
```

```python
E = np.array(E)
E
```

```python
score = 1 / (1000 * E.mean())
score
```
